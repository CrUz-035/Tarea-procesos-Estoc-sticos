1: Proceso dicotómico de Markov (Proceso telegráfico aleatorio)

La función $Y(t)$ es un proceso de Markov con rango {$-1, 1$}, que conmuta entre los dos valores a una tasa $\gamma$.

(a) Construye la ecuación maestra para el proceso dicotómico de Markov.

(b) Resuelve la ecuación maestra con la condición inicial $y(t_0) = y_0$ para obtener la densidad de probabilidad condicional $P(y, t \mid y_0, t_0)$.

> Solución

a)

Sea $x_1=-1$ y $x_2=1$

Sea $P_(x_1,t)$ la probabilidad de que $Y(t)=-1$ a un tiempo t.

Sea $P_(x_2,t)$ la probabilidad de que $Y(t)=1$ a un tiempo t.

La función cambia de -1 a 1 con una tasa $γ$.

La función cambia de 1 a -1 con una tasa $γ$.

luego, la ecuación maestra es:

$\frac{d}{dt}P(x_1, t)= -γ P_(x_1,t)+ γ P_(x_2,t)$

$\frac{d}{dt}P(x_2, t)= -γ P_(x_2,t)+ γ P_(x_1,t)$

b)

Partiendo de

$\frac{d}{dt}P(x_1, t)= -γ P_(x_1,t)+ γ P_(x_2,t)$   (1)

$\frac{d}{dt}P(x_2, t)= -γ P_(x_2,t)+ γ P_(x_1,t)$    (2)

Sumando (1) y (2), se tiene que

$\frac{d}{dt}P(x_1, t) + \frac{d}{dt}P(x_2, t)= -γ P_(x_1,t) -γ P_(x_2,t)+ γ P_(x_2,t) + γ P_(x_1,t)$

$\frac{d}{dt}(P(x_1, t) +P(x_2, t))=0,$

De aquí se tiene que $P(x_1,t)+P(x_2,t)=1$

Restando (1) y (2) se tiene

$\frac{d}{dt}P(x_1, t) - \frac{d}{dt}P(x_2, t)= -γ P_(x_1,t) +γ P_(x_2,t)+ γ P_(x_2,t) - γ P_(x_1,t)$

Reagrupando,

$\frac{d}{dt}(P(x_1, t) - P(x_2, t))+ 2γ( P_(x_1,t) - P_(x_2,t))=0$

Multiplicando por $e^{2γt}$ ambos lados de la igualdad se obtiene

$e^{2γt} \frac{d}{dt}(P(x_1, t) - P(x_2, t))+ 2γe^{2γt}( P_(x_1,t) - P_(x_2,t))=0$

$\frac{d}{dt}(e^{2γt} (P(x_1, t) - P(x_2, t)))=0$

Integrando respecto a t, se tiene que

$e^{2γt} (P(x_1, t) - P(x_2, t))=C$

Si $Y(t_0)=y_0$, nótese que $P(x_1, t_0) - P(x_2, t_0) = y_0$ para $y_0 $ $ϵ ${-1,1}.

Luego,

$e^{2γt_0}y_0=C$

Por lo tanto,

$e^{2γt} (P(x_1, t) - P(x_2, t))=e^{2γt_0}y_0$

$P(x_1, t) - P(x_2, t)=e^{2γ(t_0-t)}y_0$

Con $P(x_1, t) + P(x_2, t)=1 ⟺ P(x_1, t)=1- P(x_2, t) ⟺ P(x_2, t)=1- P(x_1, t)$

Se concluye que

$P(x_1, t | y_0,t_0) = e^{2γ(t_0-t)}y_0 +1- P(x_1, t| y_0,t_0) $

$P(x_1, t| y_0,t_0) = \frac{1}{2}(e^{2γ(t_0-t)}y_0 +1)$

De forma análoga,


$P(x_2, t| y_0,t_0) = \frac{1}{2}(-e^{2γ(t_0-t)}y_0 +1)$

De forma general

$P(y, t| y_0,t_0) = \frac{1}{2}(ye^{2γ(t_0-t)}y_0 +1)$

$P(y, t| y_0,t_0) = \frac{1}{2}(ye^{-2γ(t-t_0)}y_0 +1)$

**2: Sistema multiestable (Toggle switch)**  

Una red de "toggle switch" consiste en dos especies que se reprimen mutuamente, $r_1$ y $r_2$. Si $r_1$ es alto, la síntesis de $r_2$ es baja y, recíprocamente, si $r_2$ es alto, $r_1$ se mantiene bajo. Una red simple que describe este sistema está dada por las siguientes cuatro reacciones:

$r_1 \xrightarrow{\nu_1} r_1 + 1, \quad \nu_1 = \alpha \, g_R(r_2/\Omega)$


$r_2 \xrightarrow{\nu_2} r_2 + 1, \quad \nu_2 = \alpha \, g_R(r_1/\Omega)$

$r_1 \xrightarrow{\nu_3} r_1 - 1, \quad \nu_3 = \beta \frac{r_1}{\Omega}$

$r_2 \xrightarrow{\nu_4} r_2 - 1, \quad \nu_4 = \beta \frac{r_2}{\Omega}$

donde la función $g_R(x)$ es alta si $x$ es bajo, y baja si $x$ es alto.

**a)** Escribe la matriz de estequiometría $\mathbf{S}$ y el vector de propensiones $\vec{\nu}$ para el toggle switch. Escribe las ecuaciones deterministas de tasas de reacción y la ecuación maestra.

**b)** Supón que $g_R(x)$ toma la forma de Hill:


$g_R(x) = \frac{1}{1 + \left( \frac{x}{K_R} \right)^n}$,

donde $K_R$ mide la fuerza del represor (cuanto menor es $K_R$, menos represor se necesita para reducir $g_R$) y $n$ es la cooperatividad que determina cuán abrupta es la transición del estado alto al bajo. Realiza simulaciones estocásticas del modelo ($\alpha = 2$, $\beta = 1$, $K_R = 1$, $n = 2$) para distintos tamaños del sistema $\Omega$. ¿Observaste conmutación espontánea en las simulaciones estocásticas?



> Solución

Veamos que


S =
$\begin{array}{cc}
& \begin{array}{cccc} \nu_1 & \nu_2 & \nu_3 & \nu_4 \end{array}\\
\begin{array}{c} x_1 \\ x_2 \end{array} &
\begin{pmatrix}
1 & 0 & -1 & 0 \\
0 & 1 & 0 & -1
\end{pmatrix}
\end{array}$

es la matriz estequimétrica que indica cuánto cambian las especies cuando se lleva a cabo una reacción.

Nótese que

$\vec{\nu}_i=
\begin{pmatrix}
\nu_1\\
\nu_2\\
\nu_3\\
\nu_4
\end{pmatrix}
= \begin{pmatrix}
\alpha \, g_R(r_2/\Omega)\\
 \alpha \, g_R(r_1/\Omega)\\
 \beta \frac{r_1}{\Omega}\\
 \beta \frac{r_2}{\Omega}
\end{pmatrix}$

En el límite macroscópico, la dinámica determinista está dada por

$x_i=\frac{r_i}{Ω}$, $Ω$ el tamaño del sistema y $\frac{dx_i}{dt}=$\lim_{Ω \to ∞} SV/Ω$, con S la matriz estequimétrica y V el vector de propensiones.

Entonces

$\frac{dx_1}{dt}= \alpha \, g_R(x_2)-\beta x_1$

$\frac{dx_2}{dt}= \alpha \, g_R(x_1)-\beta x_2$.


La ecuación maestra es

$\frac{dP(r_1,r_2,t)}{dt}=\alpha \, g_R(r_2/\Omega)P(r_1-1,r_2,t)+\alpha \, g_R(r_1/\Omega)P(r_1,r_2-1,t)+\beta \frac{r_1+1}{\Omega}P(r_1+1,r_2,t)+\beta \frac{r_2+1}{\Omega}P(r_1,r_2+1,t)-[\alpha \, g_R(r_2/\Omega)+\alpha g_R(r_1/\Omega)+\beta \frac{r_1}{\Omega}+\beta \frac{r_2}{\Omega}]P(r_1,r_2,t)$

```python
#b)

import numpy as np
import math
import random as rnd
import matplotlib.pyplot as plt

#En base al problema, las condiciones son:
alpha=2.0
beta=1.0
K_r=1.0
t_0=0.0
X_0=50.0
Y_0=40.0
n=2.0
Steps=10000

#Se define la matriz estequiométrica
S=[[1.,0.,-1.,0.],[0.,1.,0.,-1.]]

#Función generadora de números aleatorios de distribución exponencial
def dist_exp(a):
	r = rnd.random()
	return -(1./a)*math.log(r)

#Se elige aleatoriamente la reacción.
def dist_reaction(ni,A):
	r = rnd.random()
	if (r < ni[0]/A):
		return 0
	elif (r < (ni[0] + ni[1])/A):
		return 1
	elif (r < (ni[0] + ni[1] + ni[2])/A):
		return 2
	elif (r <= (ni[0] + ni[1] + ni[2] + ni[3])/A):
		return 3


for omega in [100, 500, 800, 1000, 1300, 10000]:

	#Se inicializan los vectores solución y tiempo.
	Y = np.zeros([2,Steps+1])
	t = np.zeros(Steps+1)

	#Colocando condiciones iniciales
	Y[0][0]=X_0 #Sustancia 1
	Y[1][0]=Y_0 #sustancia 2

	#Algoritmo a simular
	for i in range(Steps):
		#Vector de propensiones actual.
		ni = np.array([alpha*1/(1 +((Y[1][i]/omega)/(K_r))**n),
		                alpha*1/(1 +((Y[0][i]/omega)/(K_r))**n), \
		                beta*Y[0][i]/omega, beta*Y[1][i]/omega])
		a = sum(ni) #Tasa de reacción
		tau = dist_exp(a)#El tiempo de reacción es aleatorio.
		mu = dist_reaction(ni,a)#Reacción a llevarse a cabo.
		#se actualiza el estado del sistema y avanza el tiempo
		Y[0][i+1] = Y[0][i] + S[0][mu]
		Y[1][i+1] = Y[1][i] + S[1][mu]
		t[i+1] = t[i] + tau

	# Espacio de fases
	plt.figure()
	plt.scatter(Y[0], Y[1], s=0.1)
	plt.xlabel("r1")
	plt.ylabel("r2")
	plt.title(f"Espacio de fases (Ω = {omega})")

	# Evolución temporal de r1
	plt.figure()
	plt.plot(t, Y[0])
	plt.xlabel("t")
	plt.ylabel("r1")
	plt.title(f"Evolución temporal r1 (Ω = {omega})")

	# Evolución temporal de r2
	plt.figure()
	plt.plot(t, Y[1])
	plt.xlabel("t")
	plt.ylabel("r2")
	plt.title(f"Evolución temporal r2 (Ω = {omega})")

	plt.show()
```

De acuerdo a las simulaciones vistas, las fluctuaciones en el sitema son grandes cuando el tamaño del sistema es pequeño, por lo que se observa una conmutación espontánea. A mayor es el tamaño del sistema respecto a las concentraciones, las fluctuaciones son menores.

**3: Movimiento browniano en un campo magnético**  

Una partícula browniana de masa $m$ y carga $q$ se mueve en el espacio tridimensional bajo la influencia de un campo magnético constante y homogéneo $\vec{B}$ dirigido a lo largo del eje $z$: $\vec{B} = (0, 0, B)$. El vector velocidad $\vec{V}(t) = (V_x(t), V_y(t), V_z(t))^T$ satisface la ecuación de Langevin con tres componentes:


$m \frac{d\vec{V}}{dt} = -\beta \vec{V}(t) - q \, \vec{B} \times \vec{V}(t) + \sqrt{2\beta k_B T} \, \vec{\Gamma}(t)$,

donde $\beta$ es el coeficiente de fricción y $\vec{\Gamma}(t) = (\Gamma_x(t), \Gamma_y(t), \Gamma_z(t))^T$ es un vector de ruido blanco gaussiano con componentes independientes.

**(a)** Simula la dinámica de la velocidad.

Veáse que

$m \frac{d\vec{V}}{dt} = -\beta \vec{V}(t) - q \, \vec{B} \times \vec{V}(t) + \sqrt{2\beta k_B T} \, \vec{\Gamma}(t)$⇒



$m \frac{dV_x}{dt} = -\beta V_x(t) -qBV_y(t) + \sqrt{2\beta k_B T} \, \Gamma_x(t)$

$m \frac{dV_y}{dt} = -\beta V_y(t) +qBV_x(t) + \sqrt{2\beta k_B T} \, \Gamma_y(t)$

$m \frac{dV_z}{dt} = -\beta V_z(t) + \sqrt{2\beta k_B T} \, \Gamma_z(t)$

⇒

$dV_x = \frac{-\beta V_x(t)}{m}dt-qBV_y(t)dt + \sqrt{\frac{2\beta k_B T}{m^2}} \, \Gamma_x(t)dt$

$dV_y = \frac{-\beta V_y(t)}{m}dt+qBV_x(t)dt + \sqrt{\frac{2\beta k_B T}{m^2}} \, \Gamma_y(t)dt$

$dV_z = \frac{-\beta V_z(t)}{m}dt + \sqrt{\frac{2\beta k_B T}{m^2}}\, \Gamma_z(t)dt$

Las ecuaciones anteriores son equivalentes a

$dV_x = [\frac{-\beta V_x(t)}{m}-qBV_y(t)]dt + N(0,\frac{2\beta k_B T}{m^2} dt)$

$dV_y = [\frac{-\beta V_y(t)}{m}+qBV_x(t)]dt + N(0,\frac{2\beta k_B T}{m^2}dt)$

$dV_z = \frac{-\beta V_z(t)}{m}dt + N(0,\frac{2\beta k_B T}{m^2}dt)$

donde $N(0,\frac{2\beta k_B T}{m^2}dt)$ es un muestreo de una distancia normal centrada en 0 y varianza $\frac{2\beta k_B T}{m^2}dt.$

$N(0,\frac{2\beta k_B T}{m^2})dt=N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{dt}$

Discretizando la ecuación anterior,

$ΔV_x = [\frac{-\beta V_x(t)}{m}-qBV_y(t)]Δt + N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{\Delta t}$


$\Delta V_y = [\frac{-\beta V_y(t)}{m}+qBV_x(t)]Δt + N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{\Delta t}$

$\Delta V_z = \frac{-\beta V_z(t)}{m}Δt + N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{\Delta t}$

La regla de actualización es

$ΔV_{x,i+1} = [\frac{-\beta V_{x,i}(t)}{m}-qBV_{y,i}(t)]Δt + N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{\Delta t}$


$\Delta V_{y,i+1} = [\frac{-\beta V_{y,i}(t)}{m}+qBV_{x,i}(t)]Δt + N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{\Delta t}$

$\Delta V_{z,i+1} = \frac{-\beta V_{z,i}(t)}{m}Δt + N(0,1)\sqrt{\frac{2\beta k_B T}{m^2}}\sqrt{\Delta t}$

con $t_{i+1}=t_i+\Delta t$

```python
#Condiciones iniciales
beta=0.2
k_b=1.38*10**(-23)
T=300.0
m=1.10
q=1.5
B=1.28
#Inicialización de vector y tiempo
Steps=10000
V = np.zeros((Steps+1,3))
t = np.zeros(Steps+1)
V[0]=[0,0,0]
#Algoritmo a simular
delta_t=0.001
for i in range(Steps):
  a_x=np.random.normal() #número aleatorio normal
  a_y=np.random.normal() #número aleatorio normal
  a_z=np.random.normal() #número aleatorio normal
  V[i+1,0]=V[i,0]+(-beta*V[i,0]/m-q*B*V[i,1])*delta_t+a_x*(2*beta*k_b*T/m**2)**(1/2)*(delta_t)**(1/2)#Actualización del sistema.
  V[i+1,1]=V[i,1]+(-beta*V[i,1]/m+q*B*V[i,0])*delta_t+a_y*(2*beta*k_b*T/m**2)**(1/2)*(delta_t)**(1/2)
  V[i+1,2]=V[i,2]-beta*V[i,2]/m*delta_t+a_z*(2*beta*k_b*T/m**2)**(1/2)*(delta_t)**(1/2)
  t[i+1]=t[i]+delta_t

plt.figure()
plt.plot(t, V[:,0])
plt.xlabel("t")
plt.ylabel("Vx")
plt.title("Evolución temporal de Vx")
plt.show()
plt.figure()
plt.plot(t, V[:,1])
plt.xlabel("t")
plt.ylabel("Vy")
plt.title("Evolución temporal de Vy")
plt.show()
plt.figure()
plt.plot(t, V[:,2])
plt.xlabel("t")
plt.ylabel("Vz")
plt.title("Evolución temporal de Vz")
plt.show()
```
